#### 1. 오답노트

``` C++
int main()
{
    cin >> N >> K;
    while (true)
    {
        if(N == 0)
        {
            break;
        }
        else if(N % K == 0)
        {
            N /= K;
        }
        else
        {
            N--;
        }
        ans++;
    }

    cout << ans << endl;

    
}
```

이런식으로 코딩하면 문제가 생긴다.  
두 번째 코드인 ```N /= K;``` 때문이다.

저 코드에서 N은 이렇게 변한다.

3번째 조건 실행 : 16  
2번째 조건 실행 : 4  
2번째 조건 실행 : 1 (이게 문제!)  
3번째 조건 실행 : 0  
1번째 조건 실행 : 0  

즉 4를 4로 나눠줘도 1이 된다는 것 이 부분은 맨 위의 조건을 ```N == 1```로 고쳐두면 되긴 한다. 왜냐하면 먼저 나눠주고 뺄셈을 진행하기 떄문이다.

그럼 빼고 나눠주면요? 라고 할 수 있는데 그것도 문제가 있다.

````python
N, K = map(int,input().split())
ans = 0

while True:
    if N == 1:
        break
    elif N % K != 0:
        N -= 1
    else:
        N //= K
    ans += 1
    
print(ans)
````

이것도 위에 N이 1일때로 봐야 한다.

얻어갈 것 `4를 4로 나누면 몫이 0이 아니라 1이 된다. 나눌때는 몫이 1인 경우를 생각하자!`
